```python
data_cleaned=data_cleaned.loc[data_cleaned.isnull().mean(axis=1)*100 <= 50]
```

```python
data_cleaned.head()
```

```python
columns=["PM2.5","PM10","NO","NO2","NOx","NH3","CO","SO2","O3","Benzene","Toluene"]
for col in columns:
    data_cleaned[col]=data_cleaned[col].fillna(data_cleaned[col].median())
print(data_cleaned[columns].isnull().sum())
```

```python
data_cleaned.isnull().sum()
```

```python
data_cleaned.isna()[["AQI","AQI_Bucket"]]
```

```python
data_cleaned["AQI_Bucket"].value_counts()
```

```python
buckets=["Moderate","Satisfactory","Poor","Very Poor","Good","Severe"]
max_min_value={}
for bucket in buckets:
    bucket_data=data_cleaned[data_cleaned["AQI_Bucket"]==bucket]

    min_value=bucket_data["AQI"].min()
    max_value=bucket_data["AQI"].max()
    max_min_value[bucket]={"Min AQI":min_value ,"Max AQI":max_value}
for bucket, values in max_min_value.items():
    print(f"{bucket} : \n Min_value:{values["Min AQI"]} , Max_value:{values["Max AQI"]}")
```
